```python
import pandas as pd


path = "../data/vendors_final.csv"
df = pd.read_csv(path)


print(df.shape)
df.head()
```

```python
df["text"] = (
df["name"].astype(str) + " " +
df["category"].astype(str) + " " +
df["location"].astype(str) + " " +
df["services"].astype(str)
)


print(df["text"].iloc[0])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer(stop_words="english", max_features=5000)
X = vectorizer.fit_transform(df["text"])


print(X.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity


def recommend_vendors(query, top_n=5):
    query_vec = vectorizer.transform([query])
    similarities = cosine_similarity(query_vec, X)[0]
    
    
    top_indices = similarities.argsort()[-top_n:][::-1]
    results = df.iloc[top_indices][["name", "category", "location", "services"]]
    
    
    return results
```

```python
recommend_vendors("cheap wedding venue in city center")
```
